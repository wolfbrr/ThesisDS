```python
from utils import create_facts_and_examples
import numpy as np
import pandas as pd
```

```python
N = 100
```

```python
A =  np.random.randint(2, size=N)
B =  np.random.randint(2, size=N)
C =  np.random.randint(2, size=N)
```

```python
#create artificial knowledge if A and B then Target

Target = A & B
df = pd.DataFrame({'A':A, 'B':B, "Target":Target})
create_facts_and_examples(df_=df, target='Target', 
                          predicates=['A', 'B', ], output_dir='examples/a-b')
```

```python
#create artificial knowledge if A and B and C then Target
Target = A & B & C
df = pd.DataFrame({'A':A, 'B':B, "C":C, "Target":Target})
df.to_parquet(path='examples/a-b-c' + '/df.parquet')

create_facts_and_examples(df_=df, target='Target', 
                          predicates=['A', 'B', "C"], output_dir='examples/a-b-c')
```
