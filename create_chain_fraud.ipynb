```python
from utils import create_facts_and_examples
import numpy as np
import pandas as pd
import random, os
np.random.seed(0) #Code B
```

```python
N = 30
A =  np.random.randint(2, size=N)
B =  np.random.randint(2, size=N)
C =  np.random.randint(2, size=N)
D =  np.random.randint(2, size=N)

customer = ['c'+str(i+16000) for i in range(90)]
merchant = ['m'+str(j+32000) for j in range(10)]

Target = A & B & C# & D
df = pd.DataFrame({'orig':np.random.choice(customer+merchant, size=N), 'destination':np.random.choice(merchant, size=N), 'A':A, 'B':B, "C":C, "D":D, "Target":Target})
```

```python
fromM2M = df[df['orig'].map(lambda x:x[0])=='m'].reset_index()
```

```python
fraudsters = df[df['Target']==True]
```

### finding an index where trasnactions from Merchant to merchant were performed by a fraudster and assign those transactions a flag true

```python
index_sr = pd.merge(fromM2M.reset_index(), fraudsters[['destination']].drop_duplicates(), how='inner', left_on='orig', right_on='destination')['index']
```

```python
df['Cons_orig']=0
df.loc[index_sr,'Target']=1
df.loc[index_sr,'Cons_orig']=1
df['From_orig']=1
df['To_destination']=1
```
